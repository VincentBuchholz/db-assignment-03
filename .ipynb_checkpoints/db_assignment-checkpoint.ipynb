# Connection

```python
URI = "bolt://localhost:7687"
AUTH = ("neo4j","mysecretpassword")
```

```python
driver = GraphDatabase.driver(URI, auth=AUTH)
```

```python
driver.verify_connectivity()
```

```python
def execute_write_query(query, params={}):
    with driver.session(database="neo4j") as session:
        results = session.execute_write(lambda tx: tx.run(query, params).data())
        return results
```

```python
def execute_query(query, params={}):
    with driver.session(database="neo4j") as session:
        results = session.execute_read(lambda tx: tx.run(query, params).data())
        return results
```

### Design of database and data processing:

```python
schema_viz = """
    call db.schema.visualization()
"""
```

```python
# Create an empty graph
G = nx.Graph()

# Add nodes and edges to the graph
for item in response:
    for node in item['nodes']:
        G.add_node(node['name'], label=node['name'])
    for rel in item['relationships']:
        G.add_edge(rel[0]['name'], rel[2]['name'], label=rel[1])

# Draw the graph
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edge_labels(G, pos, font_size=10)
plt.axis('off')
plt.show()
```

The schema of the database is designed as seen in the graph above. The city binds all the the nodes together, and has a relationship to all. Much of the data is stored in the relationships, e.g. one city is connected to zero or more sectors with the connection EMISSION_TARGET. The sector node has a name(Total, Buildings, etc) and the relationship contains propperties like the EmissionTarget propperty.

### Performance:
Our database is quite small, which means that we don’t have any performance problems, as all of our queries are executed quite fast, but if we got more data, we could do it better by creating indexes to make queries faster and using commands like “explain” before an query to access the execution plan that provides an insight in which nodes the query uses.

### Scaling: 
For scaling an Neo4j database it is possible to scale both horizontally and vertically, with vertically scaling we can increase/decrease the cpu and memory in the config file giving our database more resources to perform better. For horizontally scaling we are able to use causal clustering where each instance has a copy of the dataset, and write and read operations can be distributed across the cluster helping with fault tolerance, and higher loads.

### Queries:

### Can we find clusters of similiar cities:

```python
#Community detection:

proj = """
    CALL gds.graph.project('commProj', ['City', 'EmissionStatus'], {
    EMISSION:{orientation:'UNDIRECTED'}
    });
"""
execute_query(proj)
```

```python
# Setting the found communityIds
community = """
    CALL gds.graph.nodeProperty.stream('proj3','communityId', ['City'])
    YIELD nodeId, propertyValue
    WITH gds.util.asNode(nodeId) AS n, propertyValue AS communityId
    WHERE n:City
    set n.communityId = communityId
"""

execute_write_query(community)
```

```python
# how many communites are found?
count = """
    match(n:City)
    return count (DISTINCT n.communityId)
"""

execute_query(count)
```

```python
# Select a community to look at:

match = """
   MATCH (country:Country)<-[c:COUNTRY]-(n: City)-[em:EMISSION]->(emissionStatus:EmissionStatus)
    where n.communityId = 15
    RETURN n.cityName as cityName, country.countryName, em.totalEmissions as totalEmissions,  n.communityId as communityId
"""

response = execute_query(match)
```

```python
table_data = []
for item in response:
    table_data.append([
        item['cityName'],
        item['country.countryName'],
        item['totalEmissions'],
        item['communityId']
    ])

# Define headers for the table
headers = ["City Name", "Country", "Total Emissions", "Community ID"]

# Print the table using tabulate
print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

this seems to give us a mix of totally different cites...

### Do C40 cities have higher reduction targets:

```python
avg_pr_city = """
    MATCH (city:City)-[et:EMISSION_TARGET]->(sector:Sector)
    where et.percentageReductionTarget is not null
    RETURN city.c40, AVG(et.percentageReductionTarget) 
"""
```

```python
# Execute transaction
response = execute_query(avg_pr_city)
```

```python
city_labels = [entry['city.c40'] for entry in response]
avg_reduction = [entry['AVG(et.percentageReductionTarget)'] for entry in response]

plt.figure(figsize=(8, 6))
plt.bar(city_labels, avg_reduction, color=['blue', 'green'])
plt.xlabel('C40 City')
plt.ylabel('Average Percentage Reduction Target')
plt.title('Average Percentage Reduction Target for cities')
plt.ylim(0, 70)
plt.grid(axis='y')

plt.show()
```

according to our data it looks like the average percentage reductions targets for c40 cities are slightly higher than those of non c40 cities.

### Do C40 cities have lower emissions per citizen

```python
avg_pr_citizen = """
    MATCH (cityPop:CityPopulation)-[c:CITY]->(city:City)-[em:EMISSION]->(emissionStatus:EmissionStatus)
    RETURN city.c40,  AVG(em.totalEmissions/cityPop.populationAmount) as AvgEmissionsPrCitizen
"""
```

```python
# Execute transaction
response = execute_query(avg_pr_citizen)
```

```python
response
```

```python
# Extracting data
city_labels = [entry['city.c40'] for entry in response]
avg_emissions = [entry['AvgEmissionsPrCitizen'] for entry in response]

# Creating bar chart
plt.figure(figsize=(8, 6))
plt.bar(city_labels, avg_emissions, color=['red', 'blue'])
plt.xlabel('C40 City')
plt.ylabel('Average Emissions per Citizen')
plt.title('Average Emissions per Citizen')
plt.ylim(0, 15)
plt.grid(axis='y')

# Display the plot
plt.show()
```

according to our data it looks like the average emissions by citizens are infact lower for c40 cities.

### Does land area have correlation with emissions

```python
landarea_emissions="""
    MATCH (city:City)-[em:EMISSION]->(emissionStatus:EmissionStatus)
    WHERE city.landArea is not null and em.totalEmissions is not null
    RETURN city.cityName AS cityName, city.landArea AS landArea, em.totalEmissions as totalEmissions
"""
```

```python
response = execute_query(landarea_emissions)
```

```python
# Extracting data
city_names = [entry['cityName'] for entry in response]
land_areas = [entry['landArea'] for entry in response]
total_emissions = [entry['totalEmissions'] for entry in response]

plt.figure(figsize=(10, 6))
plt.scatter(land_areas, total_emissions, color='blue', marker='o')
plt.xlabel('Land Area')
plt.ylabel('Total Emissions')
plt.title('Scatter Plot of Land Area vs Total Emissions')
plt.grid(True)

# Display the plot
plt.show()
```

according to out data, cities with greater land area do not have higher emissions pr citizen.

### Does the average annual temperature of a city have a correlation with emissions

```python
temp_avg_emission = """
MATCH (city:City)-[em:EMISSION]->(emissionStatus:EmissionStatus)
WHERE city.averageAnnualTemperature IS NOT NULL and em.totalEmissions IS NOT NULL
RETURN city.cityName AS cityName, city.averageAnnualTemperature AS avgAnnualTemp, AVG(em.totalEmissions) AS avgEmissions
"""
```

```python
response = execute_query(temp_avg_emission)
```

```python

avg_annual_temps = [record['avgAnnualTemp'] for record in response]
avg_emissions = [record['avgEmissions'] for record in response]

plt.figure(figsize=(10, 6))
plt.scatter(avg_annual_temps, avg_emissions, color='blue')
plt.title('Average Emissions vs. Average Annual Temperature')
plt.xlabel('Average Annual Temperature (°C)')
plt.ylabel('Average Emissions')
plt.grid(True)
plt.show()
```

it seems that there is a slight correlation between annual temperature and emissions

###  Do cities have lower emissions than previous years

```python
i_d_status = """
    MATCH (status:EmissionStatus)<-[em:EMISSION]-(city:City)
    where status.statusName = "Increased" or status.statusName = "Decreased"
    RETURN em.reportingYear AS reportingYear, status.statusName AS statusName, COUNT(city) AS cityCount
    ORDER BY reportingYear, statusName
"""
```

```python
response = execute_query(i_d_status)
```

```python
response
```

```python

years = sorted(list(set(entry['reportingYear'] for entry in response)))
status_names = sorted(list(set(entry['statusName'] for entry in response)))
city_counts = {year: [entry['cityCount'] for entry in response if entry['reportingYear'] == year] for year in years}


plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(years))
for i, status_name in enumerate(status_names):
    plt.bar([x + i * bar_width for x in index], [city_counts[year][i] for year in years], bar_width, label=status_name)

plt.xlabel('Reporting Year')
plt.ylabel('City Count')
plt.title('City Count by Reporting Year and Status')
plt.xticks([x + bar_width * (len(status_names) - 1) / 2 for x in index], years)
plt.legend()
plt.show()
```

it seems that there is a trend of more and more cities decreasing their emissions, altough it is hard to tell when we only have data from 2016 and 2017

### What are the avg reduction targets by sector?

```python
sector_target = """
    MATCH (sector:Sector)<-[et:EMISSION_TARGET]-(city:City)
    where et.percentageReductionTarget is not null
    WITH sector, AVG(et.percentageReductionTarget) AS rt
    RETURN sector.sectorName AS sectorName, rt AS averageReductionTarget
"""
```

```python
response = execute_query(sector_target)
```

```python
sector_names = [entry['sectorName'] for entry in response]
average_reduction_targets = [entry['averageReductionTarget'] for entry in response]

plt.figure(figsize=(10, 6))
plt.bar(sector_names, average_reduction_targets, color='blue')
plt.xlabel('Sector Name')
plt.ylabel('Average Reduction Target')
plt.title('Average Reduction Target by Sector')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

### which country has the highest emissions per city 

```python
country_avg_em = """
    MATCH (city:City)-[:COUNTRY]->(country:Country)
    OPTIONAL MATCH (city)-[em:EMISSION]->(emissionStatus:EmissionStatus)
    WITH country, AVG(coalesce(em.totalEmissions, 0)) AS totalEmissions
    where totalEmissions > 0
    RETURN country.countryName AS countryName, totalEmissions
    order by totalEmissions desc
"""
```

```python
response = execute_query(country_avg_em)
```

```python
country_names = [record['countryName'] for record in response]
total_emissions = [record['totalEmissions'] for record in response]

plt.figure(figsize=(12, 6))
plt.bar(country_names, total_emissions, color='blue')
plt.title('Total Emissions by Country')
plt.xlabel('Country')
plt.ylabel('Total Emissions')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

Interestingly the highest emissions and lowest emissions both come from asian countries, being Singapore and Malaysia

### which country has the cities with the highest average reduction targets

```python
country_avg_reduction = """
    MATCH (country:Country)<-[c:COUNTRY]-(city:City)-[et:EMISSION_TARGET]->(sector:Sector)
    where et.percentageReductionTarget is not null
    WITH country, AVG(et.percentageReductionTarget) AS rt
    RETURN country.countryName AS countryName, rt AS averageReductionTarget
    order by rt desc
"""
```

```python
response = execute_query(country_avg_reduction)
```

```python
country_names = [record['countryName'] for record in response]
avg_reduction_targets = [record['averageReductionTarget'] for record in response]

plt.figure(figsize=(12, 6))
plt.bar(country_names, avg_reduction_targets, color='blue')
plt.title('Average Reduction Targets by Country')
plt.xlabel('Country')
plt.ylabel('Average Reduction Target (%)')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

### how many cities have a target of being C02 neutral in the near future

```python
c02_neutral_targets = """
    MATCH (city:City)-[et:EMISSION_TARGET]->(sector:Sector)
    where et.percentageReductionTarget = 100
    RETURN city.cityName, et.percentageReductionTarget, et.targetYear, sector.sectorName
"""
```

```python
response = execute_query(c02_neutral_targets)
```

```python
# Extracting data
table_data = [[entry['city.cityName'], entry['et.percentageReductionTarget'], entry['et.targetYear'], entry['sector.sectorName']] for entry in response]

# Table headers
headers = ['City Name', 'Percentage Reduction Target', 'Target Year', 'Sector Name']

# Print table
print(tabulate(table_data, headers=headers, tablefmt="fancy_grid"))
```
